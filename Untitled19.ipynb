```python
import pandas as pd
import numpy as np
import re
import string
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')
```

```python
true_df = pd.read_csv("True.csv")
fake_df = pd.read_csv("Fake.csv")

true_df['label'] = 1
fake_df['label'] = 0

df = pd.concat([true_df, fake_df], ignore_index=True)
df = df.sample(frac=1).reset_index(drop=True)

```

```python
stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = text.lower()
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = " ".join([word for word in text.split() if word not in stop_words])
    return text

df['text'] = df['title'] + " " + df['text']
df['text'] = df['text'].apply(clean_text)
```

```python
#classdistribution
sns.countplot(x='label', data=df)
plt.title("Class Distribution")
plt.show()
print("\n")

#Article Length Distribution
df['text_length'] = df['text'].apply(len)

plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='text_length', hue='label', bins=50, kde=True)
plt.title('Distribution of Text Length by Label')
plt.xlabel('Text Length')
plt.ylabel('Frequency')
plt.legend(title='Label', labels=['Fake', 'Real'])
plt.show()

print("\n")

#Word Count per Article
df['word_count'] = df['text'].apply(lambda x: len(x.split()))

plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='label', y='word_count')
plt.title('Boxplot of Word Count by Label')
plt.xlabel('Label')
plt.ylabel('Word Count')
plt.xticks([0, 1], ['Fake', 'Real'])
plt.show()

print("\n")

#Correlation Matrix (for Engineered Features)
# If you want to analyze correlations among numerical features
correlation_features = df[['text_length', 'word_count', 'label']]
corr = correlation_features.corr()

plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Engineered Features')
plt.show()
```

```python
vectorizer = TfidfVectorizer(max_features=5000)
X = vectorizer.fit_transform(df['text'])
y = df['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
```

```python
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
    plt.title("Confusion Matrix")
    plt.show()
```

```python
#LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
print("\n--- Logistic Regression ---")
evaluate_model(log_reg, X_test, y_test)
```

```python
#DecisionTree
dt_model = DecisionTreeClassifier(max_depth=50)
dt_model.fit(X_train, y_train)
print("\n--- Decision Tree ---")
evaluate_model(dt_model, X_test, y_test)
```

```python
#RandomForest
rf_model = RandomForestClassifier(n_estimators=100)
rf_model.fit(X_train, y_train)
print("\n--- Random Forest ---")
evaluate_model(rf_model, X_test, y_test)
```

```python
#KNN
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)
print("\n--- KNN ---")
evaluate_model(knn_model, X_test, y_test)
```
